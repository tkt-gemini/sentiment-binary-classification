## 1. Import Libraries v√† Load Model


```python
import sys
sys.path.append('..')

from app.predict import SentimentPredictor
import pandas as pd

# Load model
predictor = SentimentPredictor(model_dir='../app/models')

# Hi·ªÉn th·ªã th√¥ng tin model
model_info = predictor.get_model_info()
print(f"Model: {model_info['model_name']}")
print(f"F1-Score (Test): {model_info['f1_score']:.4f}")
print(f"Accuracy (Test): {model_info['accuracy']:.4f}")
print(f"Optimal Threshold: {model_info['optimal_threshold']:.4f}")
```

## 2. Test v·ªõi C√¢u ƒê∆°n


```python
# Test m·ªôt c√¢u
text = "Th·∫ßy gi·∫£ng b√†i r·∫•t hay v√† d·ªÖ hi·ªÉu"

result = predictor.predict_single(text)

print(f"VƒÉn b·∫£n: {result['text']}")
print(f"Sau x·ª≠ l√Ω: {result['processed']}")
print(f"\nK·∫øt qu·∫£: {result['sentiment'].upper()}")
print(f"X√°c su·∫•t T√≠ch c·ª±c: {result['prob_positive']:.1%}")
print(f"X√°c su·∫•t Ti√™u c·ª±c: {result['prob_negative']:.1%}")
```

## 3. Test v·ªõi Nhi·ªÅu C√¢u


```python
# Danh s√°ch c√¢u test
test_sentences = [
    "Th·∫ßy gi·∫£ng b√†i r·∫•t hay v√† d·ªÖ hi·ªÉu",
    "Gi·∫£ng h∆°i bu·ªìn ng·ªß, c·∫ßn c·∫£i thi·ªán th√™m",
    "C∆° s·ªü v·∫≠t ch·∫•t r·∫•t tuy·ªát v·ªùi",
    "Th∆∞·ªùng xuy√™n ƒëi mu·ªôn v√† kh√¥ng c√≥ tr√°ch nhi·ªám",
    "Gi√°o vi√™n nhi·ªát t√¨nh, lu√¥n gi√∫p ƒë·ª° sinh vi√™n",
    "B√†i gi·∫£ng kh√¥ khan, kh√¥ng sinh ƒë·ªông",
    "Ph√≤ng h·ªçc s·∫°ch s·∫Ω, tho√°ng m√°t",
    "Thi·∫øt b·ªã c≈© k·ªπ, kh√¥ng ho·∫°t ƒë·ªông t·ªët"
]

# D·ª± ƒëo√°n
results = predictor.predict_batch(test_sentences)

# Hi·ªÉn th·ªã d·∫°ng b·∫£ng
df_results = pd.DataFrame([
    {
        'VƒÉn b·∫£n': r['text'],
        'C·∫£m x√∫c': r['sentiment'],
        'X√°c su·∫•t': f"{r['probability']:.1%}",
        'T√≠ch c·ª±c': f"{r['prob_positive']:.1%}",
        'Ti√™u c·ª±c': f"{r['prob_negative']:.1%}"
    }
    for r in results
])

df_results
```

## 4. Visualize K·∫øt Qu·∫£


```python
import matplotlib.pyplot as plt
import seaborn as sns

# ƒê·∫øm s·ªë l∆∞·ª£ng m·ªói lo·∫°i c·∫£m x√∫c
sentiment_counts = df_results['C·∫£m x√∫c'].value_counts()

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(8, 5))
colors = {'positive': '#4CAF50', 'negative': '#F44336'}
ax = sns.barplot(
    x=sentiment_counts.index,
    y=sentiment_counts.values,
    hue=sentiment_counts.index,
    palette=colors
)

plt.title('Ph√¢n b·ªï C·∫£m x√∫c trong C√¢u Test', fontsize=14, pad=15)
plt.xlabel('C·∫£m x√∫c', fontsize=12)
plt.ylabel('S·ªë l∆∞·ª£ng', fontsize=12)

# Th√™m s·ªë l∆∞·ª£ng l√™n c·ªôt
for i, count in enumerate(sentiment_counts.values):
    plt.text(i, count + 0.1, str(count), ha='center', fontsize=11)

plt.tight_layout()
plt.show()

print(f"\nT·ªïng: {len(test_sentences)} c√¢u")
print(f"Positive: {sentiment_counts.get('positive', 0)} c√¢u ({sentiment_counts.get('positive', 0)/len(test_sentences)*100:.1f}%)")
print(f"Negative: {sentiment_counts.get('negative', 0)} c√¢u ({sentiment_counts.get('negative', 0)/len(test_sentences)*100:.1f}%)")
```
